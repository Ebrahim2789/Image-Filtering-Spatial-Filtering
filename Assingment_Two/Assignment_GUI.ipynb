```python
import tkinter as tk
import tkinter
import os
from tkinter.filedialog import askopenfilename, asksaveasfilename
import tkinter.messagebox
import cv2 as cv
import cv2
from __future__ import print_function
import sys
PY3 = sys.version_info[0] == 3

if PY3:
    xrange = range

import numpy as np


def basic(fn):
    window = tkinter.Tk()
    window.title("Simple Text Editor")
    window.rowconfigure(0, minsize=400, weight=1)
    window.columnconfigure(1, minsize=400, weight=1)
    
    def preprosess():
        if len(sys.argv) > 1:
            print('loading %s ...' % fn)
            img = cv.imread(fn)
            if img is None:
                print('Failed to load fn:', fn)
                sys.exit(1)

        else:
            sz = 512
            print('generating %dx%d procedural image ...' % (sz, sz))
            img = np.zeros((sz, sz), np.uint8)
            track = np.cumsum(np.random.rand(500000, 2)-0.5, axis=0)
            track = np.int32(track*10 + (sz/2, sz/2))
            cv.polylines(img, [track], 0, 255, 1, cv.LINE_AA)


        small = img
        for _i in xrange(3):
            small = cv.pyrDown(small)
        img = cv.imread(fn)
    
        def onmouse(event, x, y, flags, param):
            h, _w = img.shape[:2]
            h1, _w1 = small.shape[:2]
            x, y = 1.0*x*h/h1, 1.0*y*h/h1
            zoom = cv.getRectSubPix(img, (400, 200), (x+0.5, y+0.5))
            cv.imshow('zoom', zoom)
            if event == cv.EVENT_LBUTTONDBLCLK:
                cv.imwrite("zoom.png",zoom)
                cv.destroyAllWindows()
                main(window,"zoom.png")
#                      Add_noise()
#                      sys.exit(1)
#              cv.waitKey()
#              cv.destroyAllWindows()

        cv.imshow('preview', small)
        cv.setMouseCallback('preview', onmouse)
        print('Done')

    file = "zoom.png"
    txt_edit = tk.Text(window)
    fr_buttons = tk.Frame(window, relief=tk.RAISED, bd=2)
    
    btn_open = tk.Button(fr_buttons, text="Select Region ", command=preprosess)
    btn_save = tk.Button(fr_buttons, text="Add Noise ", command=lambda:Add_noise(window))
    btn_filter = tk.Button(fr_buttons, text="Apply Filters",command=lambda:main(window,file))

    
    # In order to display the image in a GUI, you will use the 'PhotoImage' method of Tkinter. It will an image from the directory (specified path) and store the image in a variable.
    label = tk.Label(window)

    btn_open.grid(row=0, column=0, sticky="ew", padx=5, pady=5)
    btn_save.grid(row=1, column=0, sticky="ew", padx=5)
    btn_filter.grid(row=2, column=0, sticky="ew", padx=5)

    fr_buttons.grid(row=0, column=0, sticky="ns")
    txt_edit.grid(row=0, column=1, sticky="nsew")

    icon = tkinter.PhotoImage(file = fn)
    # # Finally, to display the image you will make use of the 'Label' method and pass the 'image' variriable as a parameter and use the pack() method to display inside the GUI.
    label = tkinter.Label(window, image = icon)

    label.grid(row=0, column=1, sticky="nsew")
    # label.pack()

    window.mainloop()
```

```python
import tkinter as tk
import tkinter.messagebox
from tkinter.filedialog import askopenfilename, asksaveasfilename
import os
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from scipy.ndimage import filters
from scipy import misc
from skimage import img_as_float

ddepth=cv2.CV_16S

scale=1
k=5
sigma=1
def main(window,images):
    window.destroy()

    def Smoothing_Spatial(images):
#         Add_noise(window)
        
        if(images=="guassian.png" or images=="salt.png"):
        
            img = cv.imread(images)
            windows = tk.Frame(window, relief=tk.RAISED, bd=15)

            windows.Average=tkinter.Button(windows, text = "Average Smoothing Spatial Filter:",command=lambda:Average(window,img))
            windows.Average.grid(row=0, column=1, sticky="ew")

            windows.Median=tkinter.Button(windows, text = "Median filter:", command=lambda:median(window,img))
            windows.Median.grid(row=1, column=1, sticky="ew")

            windows.min=tkinter.Button(windows, text = "min filter:", command=lambda:median(window,img))
            windows.min.grid(row=3, column=1, sticky="ew")
            windows.max=tkinter.Button(windows, text = "max filter:", command=lambda:meane_min(window,img))
            windows.max.grid(row=2, column=1, sticky="ew")

            windows.quit = tk.Button(windows, text="Back", fg="red",command=lambda:main(window,images))
            windows.quit.grid(row=6, column=1, sticky="ew")

            windows.grid(row=0, column=0, sticky="ns")

            window.title("Smoothing Spatial Filter")
            
        else:
            tkinter.messagebox.showinfo("Alert Message", "Try Again, Invalid Operation !")

    def Sharpening_Spatial(images):
        
        if(images=="zoom.png"):
        
            
            def open_file():

                cof=txt_edit.get()
                sigma=txt_edit1.get()
                if(cof==" " and sigma ==" "):
                    tkinter.messagebox.showinfo("Alert Message", "Please Enter the cofficent Sigma!")
                    response = tkinter.messagebox.askquestion("Tricky Question", "Do you enter the parameters?")
                    print(response)
                    if response == 'yes':
                            Sharpening_Spatial(images)
                    else:
                        tkinter.messagebox.showinfo("Alert Message", "Try Again, Invalid Operation !")

                else:
                    img = cv.imread(images)

                    windows = tk.Frame(window, relief=tk.RAISED, bd=10)

                    windows.sobel=tkinter.Button(windows,text = "sobel filter",command=lambda:sobel(img,cof,sigma) )
                    windows.sobel.grid(row=0, column=1, sticky="ew")

                    windows.cany=tkinter.Button(windows,text = "cany filter", command=lambda:canny(window,img,cof,sigma))
                    windows.cany.grid(row=1, column=1, sticky="ew")

                    windows.Laplacian=tkinter.Button(windows,text = "Laplacian filter", command=lambda:Laplace(window,img,cof,sigma))
                    windows.Laplacian.grid(row=2, column=1, sticky="ew")

                    windows.Unsharp=tkinter.Button(windows,text = " Unsharp Masking and high-boost filtering", command=lambda:Unsharp_Masking(window,img,cof,sigma))
                    windows.Unsharp.grid(row=3, column=1, sticky="ew")

                    windows.Gradient= tkinter.Button(windows,text = "Gradient filter:", command=lambda:Gradiente(window,images,cof,sigma) )
                    windows.Gradient.grid(row=4, column=1, sticky="ew")

                    windows.Prewitt=tkinter.Button(windows,text = "Prewitt filter:", command=lambda:Prewitts(window,img,cof,sigma))
                    windows.Prewitt.grid(row=5, column=1, sticky="ew")

                    windows.quit = tk.Button(windows, text="Back", fg="red",command=lambda:main(window,images))
                    windows.quit.grid(row=6, column=1, sticky="ew")

                    windows.grid(row=0, column=0, sticky="ns")

                    window.title("Sharpening Spatial Filters") 

            txt_edit = tk.Entry(window,bg="blue")
            txt_edit1 = tk.Entry(window,bg="blue")
            btn_open = tk.Button(window, text="Select Filter Types", command=open_file)

            tkinter.Label(window, text = "coefficients").place(x=300, y=300)

            tkinter.Label(window, text = "Sigma").place(x=300, y=340)

            txt_edit.place(x=370, y=300)
            txt_edit1.place(x=370, y=340)       
            btn_open.place(x=300, y=370)

        else:
                tkinter.messagebox.showinfo("Alert Message", "Try Again, Invalid Operation !")
            
    def Average(window,img):
        k=5
        kernel = np.ones((k,k),np.float32)/25
        dst = cv.filter2D(img,-1,kernel)
        blur = cv.blur(img,(k,k))
        cv2.imwrite('Average.png', blur)
        files = "Average.png"
        img_av = cv.imread(files)
        Displayed(window,files)

    def median(window,img):
        k=5
        median = cv.medianBlur(img,k)
        label = tk.Label(window)
        cv2.imwrite('median.png', median)
        files = "median.png"
        Displayed(window,files)

    def meane_min(window,img):
        k=5
        img_saida_min =filters.minimum_filter(img, size = k, mode = 'constant', cval = 0)
#         plot_image(img,img_saida_min,'min')
        cv2.imwrite('meane_min.png', img_saida_min)
        files = "meane_min.png"
        Displayed(window,files)

    def meane_max(window,img):
        k=5
        img_saida_max = filters.maximum_filter(img, size = k, mode = 'constant')
        cv2.imwrite('meane_max.png', img_saida_max)
        files = "meane_max.png"
        Displayed(window,files)

    def sobel(img,cof,sigma):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        x = cv2.Sobel(gray, ddepth, 1,0, ksize=3, scale=1)
        y = cv2.Sobel(gray, ddepth, 0,1, ksize=3, scale=1)
        absx= cv2.convertScaleAbs(x)
        absy = cv2.convertScaleAbs(y)
        grad = cv2.addWeighted(absx, 0.5, absy, 0.5,0)
        cv2.imwrite('sobel.png', grad)
        files = "sobel.png"
        Displayed(window,files)

    def canny(window,img,cof,sigma):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        imagBlur=cv2.GaussianBlur(gray,(k,k),1)
        ImagCanny=cv2.Canny(imagBlur,40,40)
        cv2.imwrite('canny.png', ImagCanny)
        files = "canny.png"
        Displayed(window,files)
    def Laplace(window,img,cof,sigmaa):
        ksize = (sigma*5)|1
        smoothed = cv2.GaussianBlur(img, (ksize, ksize), sigma, sigma)
            # Apply Laplace function
        laplace = cv2.Laplacian(smoothed, ddepth, 5)
            # Converting back to uint8
        result = cv2.convertScaleAbs(laplace, (sigma+1)*0.25)
        cv2.imwrite('Laplace.png', result)
        files = "Laplace.png"
        Displayed(window,files)
        
        
    def Unsharp_Masking (window,img,cof,sigma):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
          
        imagBlur=cv2.GaussianBlur(gray,(3,3),1)
        img_mask = gray - imagBlur
        img_saida = imagBlur + (4.5 * img_mask)
    
        cv2.imwrite('Unsharp_Masking.png', img_saida)
        files = "Unsharp_Masking.png"
        Displayed(window,files)
        
      
        
    def Gradiente(window,imgs,cof,sigmaa):
        img = cv2.imread(imgs,0)
        sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
        sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
        img_saida = sobelx+sobely
        cv2.imwrite('Gradiente.png', img_saida)
        files = "Gradiente.png"
        Displayed(window,files)
        
        
    def Prewitts(window,img,cof,sigmaa):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img_gaussian = cv2.GaussianBlur(gray,(5,5),0)
        #prewitt
        kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
        kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
        img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
        img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
        Prewitt= img_prewittx + img_prewitty
        cv2.imwrite('Prewitt.png', Prewitt)
        files = "Prewitt.png"
        Displayed(window,files)
        
        
    def slidingWindowEqualization(file, winSize):
        im = cv.imread(file) 
        gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        newImg = np.zeros((gray.shape[0], gray.shape[1]))
        for row in range(gray.shape[0]-winSize+1):
            for col in range(gray.shape[1]-winSize+1):
                newImg[row:row+winSize,col:col+winSize] = cv2.equalizeHist(im[row:row+winSize,col:col+winSize])
        return newImg

 

    window = tk.Tk()
    window.title("Simple Text Editor")
    window.rowconfigure(0, minsize=400, weight=1)
    window.columnconfigure(1, minsize=400, weight=1)

    txt_edit = tk.Text(window)
    fr_buttons = tk.Frame(window, relief=tk.RAISED, bd=5)
    Smoothing_b = tk.Button(fr_buttons, text="Smoothing Spatial Filters", command=lambda:Smoothing_Spatial(images))
    Sharpening_b = tk.Button(fr_buttons, text="Sharpening Spatial Filters.", command=lambda:Sharpening_Spatial(images))
    quit = tk.Button(fr_buttons, text="Back", fg="red",command=lambda:Add_noise(window))

    # In order to display the image in a GUI, you will use the 'PhotoImage' method of Tkinter. It will an image from the directory (specified path) and store the image in a variable.
    label = tk.Label(window)


    Smoothing_b.grid(row=0, column=0, sticky="ew", padx=5, pady=5)
    Sharpening_b.grid(row=1, column=0, sticky="ew", padx=5)
    quit.grid(row=2, column=0, sticky="ew")
    
    icon = tkinter.PhotoImage(file = images)
    label = tkinter.Label(window, image = icon)
    
    fr_buttons.grid(row=0, column=0, sticky="ns")
    txt_edit.grid(row=0, column=1, sticky="nsew")
    label.grid(row=0, column=1, padx=10)

    window.mainloop()
```
